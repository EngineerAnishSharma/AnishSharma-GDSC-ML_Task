```python
# Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/sample_data/housing.csv')
df
```

```python
df.columns=[col.strip() for col in df.columns]
df.dropna(inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
df.shape
df
```

```python
from sklearn.model_selection import train_test_split
y=df["MEDV"]
x=df.drop(["MEDV"],axis='columns')
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
Model=LinearRegression()
Model.fit(x_train,y_train)
```

```python
y_pred=Model.predict(x_test)

print(f"R2 Score = {metrics.r2_score(y_test,y_pred)}")

print(f"Mean Absolute = {metrics.mean_absolute_error(y_test,y_pred)}")

print(f"Mean Squared Error = {metrics.mean_squared_error(y_test,y_pred)}")

print(f"Root Mean Squared Error = {np.sqrt(metrics.mean_squared_error(y_test,y_pred))}")
```

```python
plt.figure(figsize=(10,6))
plt.scatter(y_pred,y_test,color="blue",marker='o',alpha=0.7,label='Predictions')
plt.plot([min(y_pred),max(y_pred)],[min(y_pred),max(y_pred)],color='red',linestyle='-',linewidth=2,label='Perfect Fit')
plt.xlabel("Actual MEDV")
plt.ylabel("Predicted MEDV")
plt.title("Linear Regression:Predicted vs Actual age Expectancy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(range(len(y_test)),y_test,color="blue")
plt.scatter(range(len(y_pred)),y_pred,color="red")
plt.xlabel("Actual MEDV")
plt.ylabel("Predicted MEDV")
plt.title("Linear Regression:Predicted vs Actual age Expectancy")
plt.legend()
plt.grid(True)
plt.show()
```
